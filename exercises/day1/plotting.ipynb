```python
import numpy as np
from pylab import *
```

```python
f = open("morton_curve.dat","r")

f = [line.strip() for line in f]
f = [line.strip().split('\t') for line in f ]
data = [[float(line[0]),float(line[1])] for line in f ]
data_x = [line[0] for line in data ]
data_y = [line[1] for line in data ]

#print data

plt.scatter(data_x, data_y, color="green")
plt.plot(data_x, data_y)
plt.show()
```
